```python
#year2022, year2022fl, year2022dl are the dataframes that we will be working with 
#writing a function to create a database that an be called later as per requirements
def create_database():
    conn=psycopg2.connect("host=127.0.0.1 dbname=myfirstdb user=postgres password=great")
    conn.set_session(autocommit=True)
    cur=conn.cursor()
    
    cur.execute("drop database if exists f1tables")
    cur.execute("create database f1tables")
    conn.close()
    
    conn=psycopg2.connect("host=127.0.0.1 dbname=f1tables user=postgres password=great")
    cur=conn.cursor()
    
    return cur,conn
```

```python
cur,conn=create_database()
```

```python
year2022_table_create=("""CREATE TABLE IF NOT EXISTS year2022(
Grand_Prix VARCHAR PRIMARY KEY,
date VARCHAR,
winner VARCHAR,
car VARCHAR,
laps numeric,
time VARCHAR,
year numeric
)""")
```
